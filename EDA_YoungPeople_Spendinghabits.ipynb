# 4.1 What is the overall % interest of MEN and WOMEN in various categories

```python
#Question1 What is the overall % of interest of MEN and WOMEN in various categories 
a = spend_habit_data['Gender'][spend_habit_data['Spending_on_healthyEating'] ==5].value_counts()
b = spend_habit_data['Gender'][spend_habit_data['Spending_on_gadgets'] ==5].value_counts()
c = spend_habit_data['Gender'][spend_habit_data['Spending_on_looks'] ==5].value_counts()
d = spend_habit_data['Gender'][spend_habit_data['Entertainment_spending'] ==5].value_counts()
e = spend_habit_data['Gender'][spend_habit_data['Shopping_centres'] ==5].value_counts()
print('Num of MEN who strongly agree on spending money for Healthy Eating  ',a['male'])
print('Num of MEN who strongly agree on spending money for Gadgets         ',b['male'])
print('Num of MEN who strongly agree on spending money for Looks           ',c['male'])
print('Num of MEN who strongly agree on spending money for Entertainment   ',d['male'])
print('Num of MEN who strongly agree on spending money for Shopping Centres', e['male'])
print("_____")

print('Num of WOMEN who strongly agree on spending money for Healthy Eating   ',a['female'])
print('Num of WOMEN who strongly agree on spending money for Gadgets          ',b['female'])
print('Num of WOMEN who strongly agree on spending money for Looks            ',c['female'])
print('Num of WOMEN who strongly agree on spending money for Entertainment    ',d['female'])
print('Num of WOMEN who strongly agree on spending money for Shopping Centres ', e['female'])
labels ='healthyEating','Gadgets','Looks','Entertainment','ShoppingCentres'
values_male =  [a['male'] , b['male'] , c['male'] , d['male'],e['male'] ]
values_female =  [a['female'] , b['female'] , c['female'] , d['female'],e['female'] ]

fig, axs = plt.subplots(1,2,figsize= (15,10))

# A standard pie plot
axs[0].pie(values_male, labels=labels, autopct='%1.1f%%', shadow=True)
axs[0].set_title('MALE ')

axs[1].pie(values_female, labels=labels, autopct='%.0f%%', shadow=True,
              explode=(0, 0, 0, 0,0))
axs[1].set_title('FEMALE')

plt.show()
```

Inference <br>
Young __Women__ prefer to go to __Shopping Centres__ while young __Men__ prefer on buying __gadgets__

# 4.2 Age Range who wish to go Shopping Centres

```python
#Question2 Age Range who wish to go Shopping Centres 

as_fig = sns.FacetGrid(spend_habit_data,hue = 'Shopping_centres', aspect = 5 )
as_fig.map(sns.kdeplot , 'Age' , shade = True )
oldest = spend_habit_data['Age'].max()
as_fig.set(xlim= (0 , oldest))
as_fig.add_legend()                     # this will give the details like which line is what see right side 
plt.title('Age distribution Using Facet Grid')  # gives the title
```

Inference<br>
__18 to 22 Years__ old adults enjoy going to __Shopping Centres__<br>
Thus youths are more attracted towards shopping malls than the Middle age groups

# 4.3 Which Gender enjoys to visit shopping Centres more ?

```python
#Question3 Which Gender enjoys to visit shopping Centres more ?

#maxrate_shop = max(spend_habit_data['Shopping_centres'])
#z= int(maxrate_shop)
gen_shop = spend_habit_data.groupby(['Gender',spend_habit_data['Shopping_centres']==int(max(spend_habit_data['Shopping_centres']))] )['Gender'].count()
print(gen_shop)

# Visulaization

sns.catplot('Shopping_centres',hue = 'Gender',  kind = 'count', data =spend_habit_data,palette='Pastel1')
plt.title('Gender vs shopping centres')
```

# 4.4 Percent of People who wish to spend more on Gadgets

```python
#Question4 What % of people strongly agree to spend on Gadgets 

strong_agree_gadgets = spend_habit_data['Spending_on_gadgets'] ==5       # strongly agree 
print(spend_habit_data.groupby(['Gender',strong_agree_gadgets])['Gender'].count())

# Visualization 

f,ax = plt.subplots(1,2,figsize= (12,5))
spend_habit_data['Gender'][strong_agree_gadgets == True].value_counts().plot.pie(explode =[0,0.05],autopct= '%1.1f%%', ax= ax[0] , shadow = True)
ax[0].set_title('Strong agreement on Spending on Gadgets %') 
```

Inference<br>
__Male__ gender wishes to spend __more money__ on latest gadgets __compared to women__

# 4.12 What is average ratings received from young generations on various categories

```python
#Question12 What is average ratings received from young generations on various categories  
spend_avg_resp = pd.Series(
spend_habit_data[['Finances', 'Shopping_centres','Branded_clothing','Entertainment_spending'
                      ,'Spending_on_looks','Spending_on_gadgets','Spending_on_healthyEating']].mean()
                                        
                        )
print(spend_avg_resp) 

#visualization 

spend_avg_resp.plot(
    figsize=(5,5), kind='barh', title='Average Response Recieved',                     #horizaontal Bar , Heading 
   color=["#30a2da","#fc4f30","#e5ae38","#6d904f","#8b8b8b",'m', 'r'], xlim=(1,5))     #Color , x limiter 
```

Inference <br>
We have recieved a __Max of responses__ for __Healthy eating__ compared to other __Categories__<br> 
That is, people have given maximum responses of how willing they are to spend on eating Healthy Food
